**society column**

```python
# society (let's start the work with the column 'society')
df['society'].value_counts()
```

So we can see many catagories are there in 'society' column.

There could be some catagories, who are written with small as well as capital letters, also ratings are coming along the names.

```python
#How many unique catagories are there in 'society' column?
df['society'].value_counts().shape
```

```python
#using regular expression we are deleting ratings-part from the society names, also making it lower case
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

Now, we can see shape has been reduced

-----

**Price column**

```python
# price
df['price'].value_counts()
```

We can see Lac, Crore both. We will convert into crore only

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df.sample(5)
```

**price_per_sqft column**

```python
# price_per_sqft
df['price_per_sqft'].value_counts()
```

So, we can see rupee symbol, comma between digits and units at the last. we do not need these

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```

```python
df.head()
```

**areaWithType column** - this data is complex. Sometimes it has multiple details. So, We will process it later. It might need feature engineering.

**bedRoom column**

```python
# bedrooms
df['bedRoom'].value_counts()
```

```python
df[df['bedRoom'].isnull()]
```

So, we can see sometimes where bedroom details is = NaN, other details are also NaN. we will delete those.

```python
df = df[~df['bathroom'].isnull()]
#If we use 'bedRoom' here, 4 columns will be deleted. @ of them contains some info. So, taking 'bathroom'. we can take 'balcony' also.
```

```python
# making split based on space, then taking first value, then making type as float
df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).astype('float')
```

```python
df.head()
```

**bathroom column**

```python
# bathroom
df['bathroom'].value_counts()
```

```python
df['bathroom'].isnull().sum()
```

```python
# making split based on space, then taking first value,
#then making type as int (as there is no NaN value, we can convert as int. Previously for bedRoom column there was NaN value, we converted to float)
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype('int')
```

```python
df.head()
```

**balcony column**

```python
# balcony
df['balcony'].value_counts()
```

```python
df['balcony'].isnull().sum()
```

```python
# making split based on space, then taking first value, then replacing 'No' with '0'
#then we are leaving it as string, because we have "3+ balcony", which can mean 4,5,6 anything.
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
df.head()
```

**additionalRoom column**

```python
df['additionalRoom'].value_counts()
```

So, here we can see complex catagories are present. Feature engineering is needed. We will leave here as it is.

but in place of NaN value, we will put 'not available'.

and we will make all as lower case

```python
df['additionalRoom'].value_counts().shape
```

```python
df['additionalRoom'].isnull().sum()
```

```python
df['additionalRoom'].fillna('not available',inplace=True)
```

```python
df['additionalRoom'].isnull().sum()
```

```python
#lower case conversion
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df.head()
```

We are not doing anything for address column as of now.

**floorNum column** - it has 2 digits. 1st is which floor, 2nd is total floors


```python
df['floorNum'].value_counts()
```

```python
df['floorNum'].isnull().sum()
```

```python
# making split based on space, then taking first value,
#then replacing 'Ground' with '0', 'basement' with '-1' etc..
#then used regular expression to extract the numerical values

df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
```

```python
df['floorNum'].value_counts() #cross verify to see all are now digits or not
```
