```python
import re

# Function to extract directional words
def extract_direction(address):
    pattern = r'\b(South|North|East|West)\b'
    match = re.search(pattern, address, re.IGNORECASE)
    if match:
        return match.group()
    else:
        return None

# Convert the 'address' column to strings and apply the function to create a new column 'Direction'
df['address'] = df['address'].astype(str)  # Ensure 'address' column contains strings
df['Direction'] = df['address'].apply(extract_direction)
```
